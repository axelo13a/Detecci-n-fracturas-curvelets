# Creacion de matriz de thresholds mínimos y máximos

```python
escala = 3
filename = "IMG0004206.jpg"
img = obtain_data(filename, data)
imgc = img_cuadrada(img)
coeffs, C = curvelet_transform(imgc["img"], level = escala)
values = np.arange(95, 100, 0.2)

iou_matrix = np.zeros((len(values), len(values)))
fn_matrix = np.zeros((len(values), len(values)))
fp_matrix = np.zeros((len(values), len(values)))

for i in range(len(values)):
    for j in range(len(values)):
            tmin = threshold_value(coeffs, percentil=values[i])
            tmax = threshold_value(coeffs, percentil = values[j] )

            coeffs_t = threshold_coeffs(coeffs, tmin, tmax, mode = "hard")
            imgr = inverse_curvelet_Transform(coeffs_t, C)

            mask = threshold(imgr, 99)
            iou_val, fn, fp = iou(mask, imgc["mask"])

            iou_matrix[i,j] = iou_val
            fn_matrix[i,j] = fn
            fp_matrix[i,j] = fp

    
```

### Gráficos

#### Matriz

```python
# Crear subplots
fig, ax = plt.subplots(1, 3, figsize=(20, 6))

# Redondear los valores de "values" al primer decimal
axis = [round(x, 1) for x in values]

# Intervalo de ticks (por ejemplo, cada 5 elementos)
tick_interval = 5

# Seleccionar los índices de los ticks a mostrar (cada 5 elementos)
ticks_x = np.arange(0, len(axis), tick_interval)
ticks_y = np.arange(0, len(axis), tick_interval)

# Aplicar los ticks y las etiquetas para los tres subgráficos
for i in range(3):
    # Elegir cuál matriz mostrar (IOU, FN o FP)
    current_matrix = [iou_matrix, fn_matrix, fp_matrix][i]
    
    # Mostrar la matriz en cada subgráfico
    im = ax[i].imshow(current_matrix)
    
    # Configurar ticks y etiquetas para el eje X y Y
    ax[i].set_xticks(ticks=ticks_x)
    ax[i].set_yticks(ticks=ticks_y)

    #Configurar nombre
    ax[i].set_xlabel("Threshold mínimo")
    ax[i].set_ylabel("Threshold máximo")
    
    # Configurar las etiquetas redondeadas
    ax[i].set_xticklabels([axis[i] for i in ticks_x], rotation=90)
    ax[i].set_yticklabels([axis[i] for i in ticks_y])
    
    # Añadir barra de colores
    plt.colorbar(im, ax=ax[i])

ax[0].set_title("IOU")
ax[1].set_title("Falsos negativos")
ax[2].set_title("Falsos positivos")
# Mostrar la gráfica
plt.show()
```

#### Imagen reconstruida y máscara

```python
# Encontrar el índice del valor máximo
max_index = np.argmax(iou_matrix)

# Convertir el índice plano a índice de fila y columna
max_row, max_col = np.unravel_index(max_index, iou_matrix.shape)

th_min = 95+0.2*max_row
th_max = 95+0.2*max_col

escala = 3

filename = "IMG0004206.jpg"
img = obtain_data(filename, data)
imgc = img_cuadrada(img)

coeffs, C = curvelet_transform(imgc["img"], level = escala)

tmin = threshold_value(coeffs, percentil = th_min)
tmax = threshold_value(coeffs, percentil = th_max)


coeffs_t = threshold_coeffs(coeffs, tmin, tmax, mode = "hard")
imgr = inverse_curvelet_Transform(coeffs_t, C)
mask = threshold(imgr, 99)
iou_val, fn, fp = iou(mask, imgc["mask"])

fig, ax = plt.subplots(1,3, figsize = (10,10))
ax[0].imshow(imgc["img"], alpha = 1)
ax[0].imshow(imgc["mask"], cmap = "Reds", alpha = 0.3)
ax[0].axis("off")
ax[0].set_title("Imagen original + Máscara")
ax[1].imshow(imgr)
ax[1].imshow(imgc["mask"], cmap = "Reds", alpha = 0.2)
ax[1].axis("off")
ax[1].set_title("Imagen reconstruida")
ax[2].imshow(mask)
ax[2].imshow(imgc["mask"], cmap = "Reds", alpha = 0.5)
ax[2].axis("off")
ax[2].set_title("Reconstruida con threshold")

print(f"Valores threshold: {th_min}-{th_max}")
print(f"IOU {iou_val}, FN: {fn}, FP: {fp}")
```
